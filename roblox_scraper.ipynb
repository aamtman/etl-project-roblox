```python
#import scrapy
from splinter import Browser
from bs4 import BeautifulSoup
import requests
import os
from webdriver_manager.chrome import ChromeDriverManager

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#url to be scraped

most_engaging_url='https://www.roblox.com/games#/sortName/Curated_96'
popular_url = 'https://www.roblox.com/games#/sortName/Popular'
top_earning='https://www.roblox.com/games#/sortName/TopGrossing'
```

```python
# Retrieve page with the requests module
browser.visit(popular_url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.body.text
```

```python
print(soup.prettify())
```

```python
results = soup.find_all("li", class_="game-card game-tile")
print(results)
```

```python
# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('a', class_="game-card-link").text
       
    # Identify and return link to listing
        link = result.a['href']

        # Print results only if title, price, and link are available
        if (title and link):
            print('-------------')
            print(title)
            print(link)
    except AttributeError as e:
        print(e)
```
